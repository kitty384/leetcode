```python
from typing import List
# 让nusm2的n个元素是最大的后n位，之后插入到nums1中

# 摘录
class Solution:
    def merge(self, nums1: List[int], m: int, nums2: List[int], n: int) -> None:
        """
        Do not return anything, modify nums1 in-place instead.
        """
        p1, p2 = m - 1, n - 1
        tail = m + n - 1
        while p1 >= 0 or p2 >= 0:
            if p1 == -1:
                nums1[tail] = nums2[p2]
                p2 -= 1
            elif p2 == -1:
                nums1[tail] = nums1[p1]
                p1 -= 1
            elif nums1[p1] > nums2[p2]:
                nums1[tail] = nums1[p1]
                p1 -= 1
            else:
                nums1[tail] = nums2[p2]
                p2 -= 1
            tail -= 1
```

```python
# 逆向双指针
# 主要利用nums1后面n个空位来做调整
class Solution:
    def merge(self, nums1: List[int], m: int, nums2: List[int], n: int) -> None:
        p1=m-1
        p2=n-1
        tail=m+n-1
        
        while p1>=0 or p2>=0 :
        # 其中没有排完就不停止
        # 这里用的是or
            if p1==-1:
            # 如果nums1排完，直接把nums2复制过去
                nums1[tail]=nums2[p2]
                # tail-=1
                p2-=1
            elif p2==-1:
            # 如果nums2排完，直接把nums1复制过去
            # 且这里所有用的是elif，而不是if，不然会出现后面代码下标越界的情况
                nums1[tail]=nums1[p1]
                p1-=1
                # tail-=1
            elif nums1[p1]>nums2[p2]:
                nums1[tail]=nums1[p1]
                # tail-=1
                p1-=1
            elif nums1[p1]<=nums2[p2]:
                nums1[tail]=nums2[p2]
                p2-=1
                # tail-=1
            tail-=1
```

```python
# 双指针
# 时间换空间
class Solution:
    def merge(self, nums1: List[int], m: int, nums2: List[int], n: int) -> None:
        sort=[]
        p1=0
        p2=0
        while p1<m or p2<n:
        # 两个都排完才停止
            if p1==m and p2<n:
                sort.append(nums2[p2])
                p2+=1
            elif p1<m and p2==n:
                sort.append(nums1[p1])
                p1+=1
            elif nums1[p1]>nums2[p2]:
                sort.append(nums2[p2])
                p2+=1
            elif nums1[p1]<=nums2[p2]:
                sort.append(nums1[p1])
                p1+=1
        nums1=sort[:]
```
